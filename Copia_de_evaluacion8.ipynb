```python
#practica8
#Ejercicio1
#Escribe un programa python que pida un nùmero por teclado y que cree un diccionario cuyas claves sean desde el nùmero 1 hasta el nùmero indicado
#y los valores sean los cuadrados de las claves
#Ejemplo si se ingresa el 4 imprima el cuadrado de 1,de 2, de 3 y de 4
#crear el diccionario
cuadrado={}
numero=int(input("Ingresa un nùmero:"))
for num in range(1,numero+1):
 cuadrado[num]=num**2
for num,valor in cuadrado.items():
  print(f"{num},{valor}") 
```
